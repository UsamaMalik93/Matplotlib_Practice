## Some arguments in plt.plot()

```python
#color argument ....jUst pass first word of color withsamll letter in ''
#marker argument....Marks the datapoints with the sign you give in ''
#linestyle argument..Styles the line type we need & pass it into the string ''
#markersize argument.. it just increase the size of marker string('o') according to numerical value passed
#linewidth argument... sets the width of the line in graph according to some numerical values passed
```

```python
plt.plot(Days,temperature,color='g',marker='o',linestyle='--',markersize=6, linewidth=2)
plt.title('Next 23 Days Temperature Prediction')
plt.xlabel("Days")
plt.ylabel('T emperature')
plt.axis([0,24,30,45])#plt.axis([xmin,xmax,ymin,ymax])==>days are x axis min=0, max=23...temperature is yaxis.min=30,max=45
plt.show()
```

# Histogram chart

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
ML_students_age=np.random.randint(20,45,(100))#by this a list of 100 students ages will be printed...between 20 to 45
PY_students_age=np.random.randint(15,40,(100))#by this a list of 100 students ages will be printed...between 15 to 40
```

```python
print(ML_students_age)
print(PY_students_age)
```

```python
plt.hist(ML_students_age)
plt.title("Studenst age Studing Machine Learning",fontsize=14)
plt.xlabel("Ages",fontsize=12)
plt.ylabel("Total Students",fontsize=12)
plt.show()
```

```python
#using bins parameter we will get defined bars ----means ranges of each bar can be easily understandable
bins=[20,25,30,35,40,45]#now bins defines bar between these ranges
plt.hist(ML_students_age,bins)
plt.title("Studenst age Studing Machine Learning",fontsize=14)
plt.xlabel("Ages",fontsize=12)
plt.ylabel("Total Students",fontsize=12)
plt.show()
#now u can see how many students are between 20-25 25-30 and soo on
```

```python
#rwidth parameter --->helps to separate each bar--- means by defualt bar size is set to 1(100%) we can set it to 0.8(80%)
plt.hist(ML_students_age,bins, rwidth=0.8)
plt.title("Studenst age Studing Machine Learning",fontsize=14)
plt.xlabel("Ages",fontsize=12)
plt.ylabel("Total Students",fontsize=12)
plt.show()
```

```python
#histtype means histogram type which is by default set to bar ... we can set it to 'step' or any
plt.hist(ML_students_age,bins,rwidth=0.85,histtype='step')
plt.title("Studenst age Studing Machine Learning",fontsize=14)
plt.xlabel("Ages",fontsize=12)
plt.ylabel("Total Students",fontsize=12)
plt.show()# now see the results
```

```python
#by using oreintation parameter we can set bars to horizontal which are vertical by defualt
plt.hist(ML_students_age,bins,rwidth=0.8,histtype='bar',orientation='horizontal')
plt.title("Studenst age Studing Machine Learning",fontsize=14)
plt.xlabel("Ages",fontsize=12)
plt.ylabel("Total Students",fontsize=12)
plt.show()# see reults
```

```python
#To change color of bars we will be using color paramter
plt.hist(ML_students_age,bins,rwidth=0.85,histtype='bar',orientation='vertical',color='green')
plt.title("Studenst age Studing Machine Learning",fontsize=14)
plt.xlabel("Ages",fontsize=12)
plt.ylabel("Total Students",fontsize=12)
plt.show()
```

```python
#to label the chart we will be using label parameter 
plt.hist(ML_students_age,bins,rwidth=0.85,histtype='bar',orientation='vertical',color='r',label="ML Students")
plt.title("Studenst age Studing Machine Learning",fontsize=14)
plt.xlabel("Ages",fontsize=12)
plt.ylabel("Total Students",fontsize=12)
plt.legend(loc=2)#Note!Also use this plt.legend() to sepecify the position when giving a label otherwise label will not work
plt.show()
```
